```python
import pandas as pd
import numpy as np

import csv
import category_encoders as ce
from sklearn import metrics

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

from rfpimp import *

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #plot package
import seaborn as sns #statist graph package

import json 
import joblib
```

```python
def filter_data(df,cate,country):
    df = df[df["id"]==cate]
    df = df[df["country"] ==country]
    return df
    
```

```python
def initial_cleaning(result):
    # counting characters
    result['char_count'] = result['tags'].str.len() ## this also includes spaces
    # converting to lower case
    result['tags'] = result['tags'].apply(lambda x: " ".join(x.lower() for x in x.split()))
    # removing stopwords
    from nltk.corpus import stopwords
    stop = stopwords.words('english')
    result['tags'] = result['tags'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
        
    return(result)    
```

```python
def cleaning(df):
    df=df[df['tags_new'].map(lambda x: x.isascii())]
    df=df.drop(df[df.tags_new =='[None]'].index) 
    df= df.dropna()
    #df.drop(['country'], axis=1,inplace=True)
    return df
```

```python
def preproc(df):
    df['title'] = df['title'].str.len() ## this also includes spaces
    df['description'] = df['description'].str.len() ## this also includes spaces
    return df
```

```python
def streak(df):
    df=df.sort_values(by ='tags_new' ) 
    #df1 = df.groupby(['tags','video_id']).mean()
    df1 = df.groupby(['tags_new','video_id']).max()
    df2 = df.groupby(['tags_new','video_id']).count()
    df2.drop(['view_count', 'likes','dislikes','comment_count','title','description','Merch','Patreon'], axis=1,inplace=True)
    df2.rename(columns={'trending_date': 'streak'}, inplace=True)
    main_df=pd.concat([df1, df2], axis=1, join="inner")
    main_df.to_csv('dataset/streak_data_from_func.csv')  
    return main_df  
```

```python
def encodeX(X):
    encoder = ce.OrdinalEncoder(cols=['tags_new'])
    #encoder = ce.CountEncoder(cols=['tags'])
    encoder.fit(X)
    X = encoder.transform(X)
    return X
```

```python
def build_model(X,y):
    
    #X=encodeX(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=55)

    # Creating Random forest 
    rf = RandomForestRegressor(n_estimators=100,random_state=55,oob_score=True) 
    rf.fit(X_train, y_train)
    oob = rf.oob_score_
    n = rfnnodes(rf)
    h = np.median(rfmaxdepths(rf))
    print(f"OOB R^2 is {oob:.5f} using {n:,d} tree nodes with {h} median tree depth")

    predictions = rf.predict(X_test)
    e1 = mean_absolute_error(y_test, predictions)
    ep1 = e1*100 / y.mean()
    
    print(f" {ep1:.2f}% error for test data")
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
    print('R square error: ',r2_score(y_test, predictions))
    
    return rf,oob
```

```python
# function for feature importance
def showimp(rf, X, y):
    features = list(X.columns)
    I = importances(rf, X, y, features=features)
    print(I)
    plot_importances(I, color='#4575b4')
```

# JSON file integration

```python
def jsonmerge(df,filepath):
    # reading the json file for CANADA

    with open(filepath,'r') as f:
        json_temp= json.loads(f.read())

    # JSON file -> nested format 
    # json normalize function from pandas -> flatten it and read into data frame
    j_df = pd.json_normalize(json_temp,record_path='items')

    # Converting the 'id' extracted from the json file to type 'int'
    j_df['id']= j_df['id'].astype(int)

    # merging the category with original dataset(df)
    final= df.merge(j_df,how ='left',left_on= 'categoryId',\
                                     right_on='id').rename(columns= {'snippet.title':'category_name'})

    #final.head()
    return(final)
```

```python
filepath='dataset/CA_category_id.json'
CA_Videos_df=jsonmerge(ca_df,filepath)
CA_Videos_df.head()
```

```python
filepath='dataset/US_category_id.json'
US_Videos_df=jsonmerge(us_df,filepath)
US_Videos_df.head()
```
